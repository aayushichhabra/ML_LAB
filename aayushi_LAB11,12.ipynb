```python
import nltk
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import FreqDist

# Download required data
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')

# Sample text
text = """Machine learning is transforming the world.
It helps computers learn from data and make intelligent decisions."""

# -----------------------------
# TOKENIZATION
# -----------------------------
words = word_tokenize(text)
sentences = sent_tokenize(text)

print("Words:", words[:10])
print("Sentences:", sentences)

# -----------------------------
# STOPWORD REMOVAL
# -----------------------------
stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words if w.lower() not in stop_words]

print("\nAfter Stopword Removal:", filtered_words[:10])

# -----------------------------
# STEMMING
# -----------------------------
stemmer = PorterStemmer()
stemmed = [stemmer.stem(w) for w in filtered_words]

print("\nStemmed Words:", stemmed[:10])

# -----------------------------
# LEMMATIZATION
# -----------------------------
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(w) for w in filtered_words]

print("\nLemmatized Words:", lemmatized[:10])

# -----------------------------
# FREQUENCY DISTRIBUTION
# -----------------------------
freq = FreqDist(filtered_words)

print("\nMost Common Words:", freq.most_common(5))

# Plot
freq.plot(10)
plt.show()
```

```python
from transformers import pipeline

# Load text generation model
generator = pipeline("text-generation", model="gpt2")

# Prompt
prompt = "Machine learning is"

# Generate text
output = generator(prompt, max_length=50, num_return_sequences=1)

# Display result
print("Generated Text:\n")
print(output[0]['generated_text'])
```

```python
prompts = [
    "Artificial Intelligence will",
    "The future of technology is",
    "Data science helps in"
]

for p in prompts:
    print("\nPrompt:", p)
    print(generator(p, max_length=40)[0]['generated_text'])
```
